```r
## 1.	Use graphs (Figure above) and find shortest weighted undirected and weighted directed paths from node a to i. Find and print the sum of path weights. Plot the obtained paths on the graph.

library(igraph) 
g1 <- graph.formula(a-b, b-c, a-d, b-d, b-e, b-f, c-f, d-e, d-g, d-h, e-f, e-h, f-h, f-i, g-h, h-i) 
E(g1)$weight=c(1,4,7,7,6,2,8,5,3,9,2,4,7,4,2,1) 
V(g1)$name <- letters[1:9] 
par(mfrow=c(1, 2)) 
(mylayout=matrix(c(1,2,3,1,2,3,1,2,3, 
                  9,9,9,8,8,8,7,7,7), nrow=9, ncol=2, byrow=FALSE)) 
plot(g1, main="Undirected graph\n(task 1a)", vertex.size = 45 , edge.label = E(g1)$weight, 
  layout=mylayout)

(weighted.path <- unlist(get.shortest.paths(g1, 1, 9)$vpath) ) 
E(g1, path=weighted.path)$color <- "red" 
plot(g1, main="Weighted path\n(task 2a)", vertex.size = 25, vertex.color="white" ,  
    edge.width=(E(g1)$weight), layout=mylayout)
```

```r
## 2.	Create a random graph and find the diameter, center, radius, articulation vertices and bridge edges.
## 1)	Diametr 
library(igraph)
set.seed(41)
G <- erdos.renyi.game(n=20, p.or.m=35, type="gnm")
E(G)$color<-"black"
V(G)[get.diameter(G)]$color <- "green"
E(G, path = get.diameter(G))$color <- "green"


V(G)[radius(G)]$color <- "red"
E(G, path = radius(G))$color <- "red"
plot(G, layout = layout.fruchterman.reingold)

```

```r
# Articulation (cut) vertices and bridge edges (red)
library(igraph)
set.seed(41)
G <- erdos.renyi.game(n=20, p.or.m=35, type="gnm")
V(G)$color <- "black"
E(G)$color <- "black"
V(G)$color[ articulation.points(G) ] <- "red"
num_comp <- length( decompose.graph(G) )
for (i in 1:length(E(G))) {
G_sub <- delete.edges(G, i)
if ( length( decompose.graph(G_sub) ) > num_comp ) E(G)$color[i] <- "red"
}
plot(G, layout = layout.fruchterman.reingold, vertex.size = 15, vertex.color= V(G)$color,
vertex.label.color = "white" )

```

```r
# 3.	Create a random graph and plot colored friends and friends of friends.
library(igraph)
set.seed(41)
g <- erdos.renyi.game(n=20, p.or.m=35,
type="gnm")
f1 <- unlist(neighborhood(g, 1, nodes=V(g)[14]))
f2 <- unlist(neighborhood(g, 2, nodes=V(g)[14]))
V(g)$color <- "white"
V(g)[f1[1]]$color <- "blue"
V(g)[f1[-1]]$color <- "green"
V(g)[setdiff(f2,f1)]$color <- "red"
plot(g, layout = layout.fruchterman.reingold)
```
